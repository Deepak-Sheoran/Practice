```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

## distplot for univariate visualisation

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=True)
```

```python
sns.displot(tips['total_bill'],kind='kde')
```

```python
sns.displot(tips['total_bill'],kind='ec')
```

## joint plot for bivaraite viualisation

```python
sns.jointplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.rugplot(tips,x='total_bill',y='tip')
```

```python
sns.rugplot(tips)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips)
```
