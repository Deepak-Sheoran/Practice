```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head(6)
```

# Bar Plot

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

So males have more standard deviation in their total bills than the females.

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.count_nonzero)
```

# Count Plot

```python
sns.countplot(tips,x='sex')
```

```python
sns.countplot(tips,y='sex')
```

# Box and Violin Plots

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(tips,x='day',y='total_bill')
```

```python
sns.violinplot(tips,x='day',y='total_bill',hue='smoker',split=True)
```

```python
sns.violinplot(tips,y='day',x='total_bill',hue='smoker',split=True)
```

# Stripplot

```python
sns.stripplot(tips,x='day',y='total_bill',hue='sex')
```

# Swarmplot

```python
sns.swarmplot(tips,x='day',y='total_bill',hue='day')
```

```python
sns.violinplot(tips,x='day',y='total_bill')
sns.swarmplot(tips,x='day',y='total_bill',color='black')
```
