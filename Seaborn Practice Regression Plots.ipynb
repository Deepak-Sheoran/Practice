```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(tips,x='total_bill',y='tip')
```

```python
sns.lmplot(tips,x='total_bill',y='tip',hue='sex',markers=['o','v'])
```

```python
sns.lmplot(tips,x='total_bill',y='tip',col='sex',row='time')
```
