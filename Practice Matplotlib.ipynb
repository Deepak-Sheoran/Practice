```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure(figsize=(1,1))
axes=fig.add_axes([0,0,.5,.5])
```

```python
x=np.linspace(1,5,10)
y=x**2
plt.plot(x,y,'r.')
plt.plot(y,x,'y--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('The Title')
```

```python
# Function Method
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'g--')
```

```python
# Object Oriented Method
fig=plt.figure()
axes=fig.add_axes([1,1,0.5,0.5])
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,0.8,0.8])
axes1.plot(x,y,'g.-')
axes2=fig.add_axes([0.1,0.4,0.3,0.3])
axes2.plot(y,x,'r--')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('Title')

axes2.set_xlabel('x Axis')
axes2.set_ylabel('y Axis')
axes2.set_title('title')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X Squared')
axes.plot(x,x**3,label='X Cubed')
axes.legend()
```
