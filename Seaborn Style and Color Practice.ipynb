```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('dark')
#sns.set_style('ticks')
sns.countplot(tips,x='sex').set_title('Count')
#ax.set_title('Count')
```

```python
plt.figure(figsize=(2,2))
sns.countplot(tips,x='sex').set_title('Count')
```
